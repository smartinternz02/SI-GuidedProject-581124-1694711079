## 8. Split into Independent & Dependent Variables

```python
y = df['quality']
y
```

```python
X =df.drop(columns =['quality'],axis =1)
X.head()
```

## 8. Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler
scale =MinMaxScaler()
```

```python
X_scaled= pd.DataFrame(scale.fit_transform(X),columns =X.columns)
X_scaled.head()
```

## 9. Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=10)
```

```python
from sklearn import metrics
```

```python
# r2 score
print(metrics.r2_score(y_test,pred1))
print(metrics.r2_score(y_test,pred2))
```

```python
# MSE
print(metrics.mean_squared_error(y_test,pred1))
print(metrics.mean_squared_error(y_test,pred2))
```
